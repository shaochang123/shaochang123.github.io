```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# 
file_path_template = r'C:\Users\15057\Desktop\create excel\esp32 data\\\{}.xlsx'

# 
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 20))  # 52
axes = axes.flatten()  # 

# 10
for i in range(1, 11):
    file_path = file_path_template.format(i)  # 
    
    try:
        # Excel
        df = pd.read_excel(file_path)
        
        # EMG_Raw
        emg_raw = df['EMG_Raw']
        
        # 
        ax = axes[i-1]
        ax.plot(emg_raw)
        ax.set_title(f'EMG_Raw Data -  {i}')
        ax.set_xlabel('Sample Index')
        ax.set_ylabel('EMG_Raw Value')
        ax.grid(True)
        
    except Exception as e:
        print(f" {file_path}: {e}")

# 
plt.tight_layout()

# 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# 
file_path_template = r'C:\Users\15057\Desktop\create excel\esp32 data\\\{}.xlsx'

# 
fig, ax = plt.subplots(figsize=(12, 8))

# 
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'purple', 'pink']

# 10
for i in range(1, 11):
    file_path = file_path_template.format(i)  # 
    
    try:
        # Excel
        df = pd.read_excel(file_path)
        
        # EMG_Raw
        emg_raw = df['EMG_Raw']
        
        # 
        ax.plot(emg_raw, label=f' {i}', color=colors[i-1])
        
    except Exception as e:
        print(f" {file_path}: {e}")

# 
ax.set_title('EMG_Raw Data from 10 Files')
ax.set_xlabel('Sample Index')
ax.set_ylabel('EMG_Raw Value')

ax.set_ylim(540, 620)
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.show()
```
