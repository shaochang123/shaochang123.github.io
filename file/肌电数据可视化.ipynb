```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# 文件路径模板
file_path_template = r'C:\Users\15057\Desktop\create excel\esp32 data\李\握拳（李）\握拳（李）{}.xlsx'

# 创建一个图表和子图
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 20))  # 5行2列的子图布局
axes = axes.flatten()  # 将子图数组展平，方便循环

# 遍历10个文件
for i in range(1, 11):
    file_path = file_path_template.format(i)  # 动态生成文件路径
    
    try:
        # 读取Excel文件
        df = pd.read_excel(file_path)
        
        # 提取EMG_Raw数据
        emg_raw = df['EMG_Raw']
        
        # 绘制折线图
        ax = axes[i-1]
        ax.plot(emg_raw)
        ax.set_title(f'EMG_Raw Data - 文件 {i}')
        ax.set_xlabel('Sample Index')
        ax.set_ylabel('EMG_Raw Value')
        ax.grid(True)
        
    except Exception as e:
        print(f"无法读取或处理文件 {file_path}: {e}")

# 调整子图间距
plt.tight_layout()

# 显示图表
plt.show()
```
